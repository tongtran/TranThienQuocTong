```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df_train = pd.read_csv('train.csv', na_values="?")
df_test = pd.read_csv('test.csv', na_values="?")
df_train
```

```python
workclass_ = {'Private':0, 'Self-emp-not-inc':1, 
                  'Self-emp-inc':2, 'Federal-gov':3, 
                  'Local-gov':4, 'State-gov':5, 
                  'Without-pay':6, 'Never-worked':7}
df_train.workclass=df_train.workclass.replace(workclass_)
df_test.workclass=df_test.workclass.replace(workclass_)

education_ = {'Preschool':0,'1st-4th':1,'5th-6th':2,
              '7th-8th':3,'9th':4,'10th':5,'11th':6,
              '12th':7,'HS-grad':8,'Some-college':9,'Assoc-voc':10,
              'Assoc-acdm':11,'Bachelors':12,'Masters':13,'Prof-school':14,'Doctorate':15}
df_train.education=df_train.education.replace(education_)
df_test.education=df_test.education.replace(education_)
```

```python
X_train = df_train.iloc[:, :-1].values
y_train = df_train.iloc[:, -1:].values
X_test  = df_test.iloc[:, :-1].values
y_test = df_test.iloc[:, -1:].values
```

```python
from sklearn.impute import SimpleImputer

workclass_impute = SimpleImputer(np.nan, strategy="most_frequent")
X_train[:, 1] = workclass_impute.fit_transform(X_train[:, 1].reshape(-1, 1)).ravel()
X_train[:, -1] = workclass_impute.fit_transform(X_train[:, -1].reshape(-1, 1)).ravel()

X_test[:, 1] = workclass_impute.fit_transform(X_test[:, 1].reshape(-1, 1)).ravel()
X_test[:, -1] = workclass_impute.fit_transform(X_test[:, -1].reshape(-1, 1)).ravel()
```

```python
from category_encoders import CountEncoder
countEncoder = CountEncoder()
marital = countEncoder.fit_transform(X_train[:,5])
X_train[:, 5] = marital.values.reshape(1,-1)
occupation = countEncoder.fit_transform(X_train[:,6])
X_train[:, 6] = occupation.values.reshape(1,-1)
relationship = countEncoder.fit_transform(X_train[:,7])
X_train[:, 7] = relationship.values.reshape(1,-1)
race = countEncoder.fit_transform(X_train[:,8])
X_train[:, 8] = race.values.reshape(1,-1)
native = countEncoder.fit_transform(X_train[:,-1])
X_train[:, -1] = native.values.reshape(1,-1)


marital = countEncoder.fit_transform(X_test[:,5])
X_test[:, 5] = marital.values.reshape(1,-1)
occupation = countEncoder.fit_transform(X_test[:,6])
X_test[:, 6] = occupation.values.reshape(1,-1)
relationship = countEncoder.fit_transform(X_test[:,7])
X_test[:, 7] = relationship.values.reshape(1,-1)
race = countEncoder.fit_transform(X_test[:,8])
X_test[:, 8] = race.values.reshape(1,-1)
native = countEncoder.fit_transform(X_test[:,-1])
X_test[:, -1] = native.values.reshape(1,-1)


```

```python
from sklearn.preprocessing import LabelEncoder
X_label_encoder = LabelEncoder()
X_train[:, 9] = X_label_encoder.fit_transform(X_train[:, 9])

X_test[:, 9] = X_label_encoder.fit_transform(X_test[:, 9])
```
